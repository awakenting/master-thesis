```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat = {'nstartles':[129, 43, 161, 56, 184, 57, 192, 53],
       'condition':['iso', 'social', 'iso', 'social', 'iso', 'social', 'iso', 'social'],
       'amp':[0.2, 0.2, 0.4, 0.4, 0.6, 0.6, 0.8, 0.8]}
```

```python
df = pd.DataFrame(dat)
```

```python
#sns.set()
sns.factorplot(x='amp', y='nstartles', hue='condition', data=df, kind='bar', palette='muted', )
plt.show()
```
