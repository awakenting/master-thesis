```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)
xx, yy = np.meshgrid(x, y, sparse=True)
coeff1 = 1
coeff2 = np.sqrt(4 - coeff1**2)
z = np.sin((coeff1*xx + coeff2*yy))
h = plt.contourf(x,y,z)
```

```python
# Take the 2-dimensional DFT and centre the frequencies
image = z
ftimage = np.fft.fft2(image)
ftimage = np.fft.fftshift(ftimage)
iftimage = np.fft.ifft2(np.fft.fftshift(ftimage))

plt.figure(figsize=(6,6))
plt.imshow(z)
plt.title('original image')
plt.figure(figsize=(6,6))
plt.imshow(np.abs(ftimage))
plt.title('fourier transform')
plt.figure(figsize=(6,6))
plt.imshow(np.real(iftimage))
plt.title('inverse fourier transform')
plt.show()
```

```python
freqs = np.fft.fftshift(np.fft.fftfreq(100))
```

```python
max_idx = np.unravel_index(np.argmax(np.abs(ftimage)), ftimage.shape)
```

```python
print(freqs[max_idx[0]], freqs[max_idx[1]])
```
